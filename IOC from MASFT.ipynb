````python
import os
import json
import re
from google import genai
from google.genai import types

# =======================
# File Paths and Gemini Configuration
# =======================
rag_prompts_dir = "rag_prompts"     # Directory containing prompt text files
output_responses_file = "gemini_responses.json"  # File to save responses

# --- Gemini LLM Configuration ---
API_KEY = "AIzaSyCjbGRnG3XdvIeUTKnwZ1HgS0sSYGg-t5E"  # Replace with your actual API key
MODEL = "gemini-1.5-flash"  # Using Gemini 1.5 Flash

client = genai.Client(api_key=API_KEY)

# --- Updated system instruction ---
system_instruction = (
    "You are a component of a Retrieval-Augmented Generation (RAG) system designed to analyze LinkedIn profiles to answer user queries. "
    "The system processes each query by:\n"
    "  1. Decomposing the query into key nodes (entities) and relations (extracted from LinkedIn profile data).\n"
    "  2. Retrieving an ordered information sequence by traversing a knowledge graph.\n"
    "  3. Fetching the original chunk texts corresponding to the information retrieved.\n"
    "Your task is to follow the instructions to answer the query from the [RETRIEVED SEQUENCE] & [UNIQUE CHUNK TEXTS]. "
    "The provided context with [RETRIEVED SEQUENCE] & [UNIQUE CHUNK TEXTS] is always complete in nature and the answer must be provided using the same context.\n"
    "Below is the complete context provided for the query."
)

config = types.GenerateContentConfig(
    system_instruction=system_instruction,
    max_output_tokens=8000,
    temperature=0.1
)

# =======================
# Process RAG Prompts: Load, Call Gemini, and Save Responses
# =======================
responses = {}  # Dictionary to hold query: response_text

for filename in os.listdir(rag_prompts_dir):
    # Expect filenames like {safe_query}_prompt.txt
    if not filename.endswith("_prompt.txt"):
        continue
    filepath = os.path.join(rag_prompts_dir, filename)
    with open(filepath, "r", encoding="utf-8") as f:
        prompt_text = f.read()
    
    # Derive query text from filename (replace underscores with spaces, remove _prompt.txt)
    safe_query = filename.replace("_prompt.txt", "")
    query_text = safe_query.replace("_", " ")
    
    # Call the Gemini LLM with the updated system instruction.
    try:
        response = client.models.generate_content(
            model=MODEL,
            contents=[prompt_text],
            config=config
        )
        # Store and display the response.
        responses[query_text] = response.text
        print(f"Response for query: {query_text}\n{response.text}\n{'-'*40}")
    except Exception as e:
        print(f"Error processing query '{query_text}': {e}")
        responses[query_text] = f"Error: {e}"

# Save all responses to a JSON file.
with open(output_responses_file, "w", encoding="utf-8") as f:
    json.dump(responses, f, indent=2)

print(f"All Gemini LLM responses saved to: {os.path.realpath(output_responses_file)}")

# =======================
# Verification Using MASFT Failure Taxonomy Framework
# =======================
# Build a verification prompt that includes all outputs.
compiled_outputs = ""
for query, response_text in responses.items():
    compiled_outputs += f"Query: {query}\nResponse:\n{response_text}\n---\n"

verification_prompt = f"""
You are a verifying agent tasked with reviewing outputs from a multi-agent system. Use the following Multi-Agent LLM System Failure Taxonomy framework to assess the outputs and flag potential issues using the definitions provided:

Failure Categories:
• Task Verification (Quality Control) – Ensuring outputs adhere to the intended task specifications.
• Poor Specification (System Design) – Instances where system instructions or role definitions are unclear or incomplete.
• Inter-Agent Misalignment (Agent Coordination) – Issues in the coordination and communication between agents.

Failure Modes and Definitions:
1.1 Disobey Task Specification (11.5%): The output does not follow the explicit task instructions.
1.2 Disobey Role Specification (6.54%): The output shows agents performing tasks outside their defined roles.
1.3 Step Repetition (7.59%): The output redundantly repeats steps or information unnecessarily.
1.4 Loss of Conversation History (9.16%): Critical context is omitted due to missing conversation history.
1.5 Unaware of Termination Conditions (15.2%): The output fails to recognize when the task should be completed.
2.1 Conversation Reset (2.36%): The system resets or loses context during interactions.
2.2 Fail to Ask for Clarification (5.50%): Agents do not seek clarification when instructions are ambiguous.
2.3 Task Derailment (6.02%): The discussion diverges from the original task.
2.4 Information Withholding (8.64%): Relevant information is omitted in the output.
2.5 Ignored Other Agent’s Input (13.61%): Agents do not incorporate or acknowledge contributions from others.
2.6 Reasoning-Action Mismatch (1.57%): The reasoning process does not align with the subsequent action.
3.1 Premature Termination (13.61%): The task ends before all requirements are met.
3.2 No or Incomplete Verification (4.71%): The output is not properly checked against the task specifications.
3.3 Incorrect Verification (2.09%): The output is verified against incorrect criteria.

The outputs to verify include:
---
{compiled_outputs}
---
Please review the above outputs and generate a detailed verification report summarizing any detected issues, inconsistencies, or areas for improvement. Include any Indicators of Compromise (IOC) where applicable.
"""

print("\nVerification Prompt:\n", verification_prompt)

# Call Gemini API to generate the verification report.
try:
    verification_response = client.models.generate_content(
        model=MODEL,
        contents=[verification_prompt],
        config=types.GenerateContentConfig(
            system_instruction="You are verifying RAG system outputs using the provided Failure Taxonomy. Evaluate responses based on adherence to task instructions, specification quality, and inter-agent coordination.",
            max_output_tokens=200,
            temperature=0.1
        )
    )
    verification_report = verification_response.text.strip() if verification_response and hasattr(verification_response, "text") else ""
    print("\nVerification Report:\n")
    print(verification_report)
except Exception as e:
    print("Error generating verification report:", e)
````
